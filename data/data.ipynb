```python
import yfinance as yf
import pandas as pd
import pickle
import requests
import os
```

# Importing data from Yfinance

```python
#initializing counters
downloaded = 0
skipped = 0
#initializing tickers lists
tickers_skipped = []
tickers_downloaded = []
#loop to scrap Ibovespa (brazilian index) constituints for each quarter
for ano in range(2003, 2021):
  for quarter in range(1, 4):
    url = f"https://raw.githubusercontent.com/igor17400/IBOV-HCI/main/historic_composition/{ano}_{quarter}Q.csv"
    tickers = pd.read_csv(url)["symbol"].to_list()
    for ticker in tickers:
      if ticker not in tickers_downloaded:
        print(f"Downloading {ticker}...")
        df = yf.download(str(ticker) + ".SA")[["Open", "Adj Close"]]
        if list(yf.shared._ERRORS.keys()):
          print(f"{ticker} not found, skipping...")
          if ticker not in tickers_skipped:
            skipped += 1
            tickers_skipped.append(ticker)
        else:
          df.to_csv(f"data/raw_data/{ticker}.csv")
          tickers_downloaded.append(ticker)
          downloaded += 1
    tickers.clear()  
```

```python
print("Downloaded completed")
print(f"Data downloaded for {downloaded} securities")
print(f"Tickers skipped: {tickers_skipped}")
len(tickers_skipped)
```

# Importing data remaining from fundamentus (https://www.fundamentus.com.br/)

```python
# creating function to scrap data from Fundamentus website
def fund_data(ticker):
    # defining url and headers to request data
    url = f"https://fundamentus.com.br/amline/cot_hist.php?papel={ticker}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0",
        "Referer": "https://fundamentus.com.br/cotacoes.php?"
    }
    # requesting data from website
    response = requests.request("GET", url, headers=headers)
    # creating empty lists to store ticker data
    index = []
    adj_close = []
    # loop trough each data row
    for i in response.text.split("\n")[1:-2]:
        # add index datetime data
        index.append(pd.to_datetime(i.split(",")[1][1:], unit="ms"))
        # add Adj Close data 
        adj_close.append(float(i.split(",")[2][:-2]))
    # create dataframe from data lists (index and Adj Close)
    df = pd.DataFrame(adj_close, index=index, columns=["Adj Close"])
    # check if dataframe is not empty
    if not df.empty:
        # save data to a csv file
        df.to_csv(f"data/raw_data_fund/{ticker}.csv")
    else:
        print(ticker)
    
    return
```

```python
# scrapping data and printing missing tickers
for ticker in tickers_skipped:
  fund_data(ticker)
```

# Checking if there is missing data

```python
# defining os path for each data source
directory = os.fsencode("data/raw_data")
directory_2 = os.fsencode("data/raw_data_fund")

# creating empty lists to store ticker info for each data source
raw_data = []
raw_data_fund = []

# loop os path for yfinance data source
for file in os.listdir(directory):
    ticker = os.fsdecode(file).replace(".csv", "")
    # saving ticker data name found in directory
    raw_data.append(ticker)

# loop os path for fundamentus data source
for file in os.listdir(directory_2):
    ticker = os.fsdecode(file).replace(".csv", "")
    # saving ticker data name found in directory
    raw_data_fund.append(ticker)
```

```python
# creating missing data ticker name list
missing_data = []

# loop trough possible missing data tickers
for ticker in list(set(tickers_downloaded + tickers_skipped) - set(raw_data + raw_data_fund)):
    # checking if ticker missing is not in yfinance source directory (raw_data) or fundamentus source directory (raw_data_fund)
    if not [ str(x).replace("b'", "").replace(".csv'", "") for x in os.listdir(directory)] or not [ str(x).replace("b'", "").replace(".csv'", "") for x in os.listdir(directory_2)]:
        # saving missing ticker in list
        missing_data.append(ticker)
```
