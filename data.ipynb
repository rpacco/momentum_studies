```python
import yfinance as yf
import pandas as pd
import pickle
import requests
import os
```

# Importing data from Yfinance

```python
#initializing counters
downloaded = 0
skipped = 0
#initializing tickers lists
tickers_skipped = []
tickers_downloaded = []
#loop to scrap Ibovespa (brazilian index) constituints for each quarter
for ano in range(2003, 2021):
  for quarter in range(1, 4):
    url = f"https://raw.githubusercontent.com/igor17400/IBOV-HCI/main/historic_composition/{ano}_{quarter}Q.csv"
    tickers = pd.read_csv(url)["symbol"].to_list()
    for ticker in tickers:
      if ticker not in tickers_downloaded:
        print(f"Downloading {ticker}...")
        df = yf.download(str(ticker) + ".SA")[["Open", "Adj Close"]]
        if list(yf.shared._ERRORS.keys()):
          print(f"{ticker} not found, skipping...")
          if ticker not in tickers_skipped:
            skipped += 1
            tickers_skipped.append(ticker)
        else:
          df.to_csv(f"data/raw_data/{ticker}.csv")
          tickers_downloaded.append(ticker)
          downloaded += 1
    tickers.clear()  
```

```python
print("Downloaded completed")
print(f"Data downloaded for {downloaded} securities")
print(f"Tickers skipped: {tickers_skipped}")
len(tickers_skipped)
```

# Updating tickers downloaded and tickers skipped list

```python
for ticker in updated_tickers:
      if ticker not in tickers_downloaded:
        print(f"Downloading {ticker}...")
        df = yf.download(str(ticker) + ".SA")[["Open", "Adj Close"]]
        if list(yf.shared._ERRORS.keys()):
          print(f"{ticker} not found, skipping...")
          if ticker not in tickers_skipped:
            skipped += 1
            tickers_skipped.append(ticker)
        else:
          df.to_csv(f"data/raw_data/{ticker}.csv")
          tickers_downloaded.append(ticker)
          tickers_skipped.remove(list(tickers_changes.keys())[list(tickers_changes.values()).index(ticker)])
          downloaded += 1
```

# Importing data remaining from fundamentus (https://www.fundamentus.com.br/)

```python
def fund_data(ticker):
    url = f"https://fundamentus.com.br/amline/cot_hist.php?papel={ticker}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0",
        "Referer": "https://fundamentus.com.br/cotacoes.php?"
    }
    response = requests.request("GET", url, headers=headers)
    index = []
    adj_close = []
    for i in response.text.split("\n")[1:-2]:
        index.append(pd.to_datetime(i.split(",")[1][1:], unit="ms"))
        adj_close.append(float(i.split(",")[2][:-2]))
    df = pd.DataFrame(adj_close, index=index, columns=["Adj Close"])
    if not df.empty:
        df.to_csv(f"data/raw_data_fund/{ticker}.csv")
    else:
        print(ticker)
    
    return      
```

```python
for ticker in tickers_skipped:
  fund_data(ticker)
```
