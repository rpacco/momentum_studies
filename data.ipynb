```python
import yfinance as yf
import pandas as pd
import pickle
import requests
```

# Importing data from Yfinance

```python
#initializing counters
downloaded = 0
skipped = 0
#initializing tickers lists
tickers_skipped = []
tickers_downloaded = []
#loop to scrap Ibovespa (brazilian index) constituints for each quarter
for ano in range(2003, 2021):
  for quarter in range(1, 4):
    url = f"https://raw.githubusercontent.com/igor17400/IBOV-HCI/main/historic_composition/{ano}_{quarter}Q.csv"
    tickers = pd.read_csv(url)["symbol"].to_list()
    for ticker in tickers:
      if ticker not in tickers_downloaded:
        print(f"Downloading {ticker}...")
        df = yf.download(str(ticker) + ".SA")[["Open", "Adj Close"]]
        if list(yf.shared._ERRORS.keys()):
          print(f"{ticker} not found, skipping...")
          if ticker not in tickers_skipped:
            skipped += 1
            tickers_skipped.append(ticker)
        else:
          df.to_csv(f"data/raw_data/{ticker}.csv")
          tickers_downloaded.append(ticker)
          downloaded += 1
    tickers.clear()  
```

```python
print("Downloaded completed")
print(f"Data downloaded for {downloaded} securities")
print(f"{tickers_skipped} tickers skipped")
len(tickers_skipped)
```

# Saving pickle file for downloaded tickers list

```python
import pickle

with open("data/tickers_downloaded.pkl", "wb") as file:
    pickle.dump(tickers_downloaded, file)
```

# Considering tickers changes during this period (1Q 2003 to 3Q 2021)

```python
# AMBV4 changed ticker to ABEV3
# https://www.infomoney.com.br/negocios/ambev-passa-a-ter-apenas-acoes-on-a-partir-da-proxima-sessao-veja-o-que-muda/

# ITAU4 changed ticker to ITUB4
# https://www.terra.com.br/economia/bovespa-itau-muda-codigos-de-negociacao-a-partir-do-dia-20,909117a7adc4b310VgnCLD200000bbcceb0aRCRD.html

# NATU3 changed ticker to NTCO3
# https://www.suno.com.br/noticias/natura-muda-ticker-ntco3/

# BVMF3 changed ticker to B3SA3
# https://exame.com/invest/mercados/b3-muda-codigo-de-suas-acoes-na-bolsa-a-partir-de-hoje/

# TIMP3 changed ticker to TIMS3
# https://www.suno.com.br/noticias/tim-tims3-timp3-codigo-ticker-b3-nyse-acoes/

# ESTC3 changed ticker to YDUQ3
# https://www.infomoney.com.br/negocios/estacio-muda-nome-da-marca-para-yduqs-e-abre-nova-fase-de-crescimento/

# KROT3 changed ticker to COGN3
# https://www.infomoney.com.br/negocios/kroton-anuncia-mudanca-de-nome-e-criacao-de-holding-com-quatro-marcas/

# RUMO3 changed ticker to RAIL3
# https://financenews.com.br/2017/03/rumo3-vai-virar-rail3/

# BRDT3 changed ticker to VIBR3
# https://www.suno.com.br/noticias/vibra-ex-br-mudara-ticker-brdt3-vbbr3/

# VVAR3 changed ticker to VIIA3
# https://www.infomoney.com.br/mercados/via-muda-de-codigo-na-bolsa-a-partir-do-dia-16-e-nao-usara-mais-o-nome-via-varejo/

tickers_changes = {
    "AMBV4": "ABEV3",
    "ITAU4": "ITUB4",
    "NATU3": "NTCO3",
    "BVMF3": "B3SA3",
    "TIMP3": "TIMS3",
    "ESTC3": "YDUQ3",
    "KROT3": "COGN3",
    "RUMO3": "RAIL3",
    "BRDT3": "VBBR3",
    "VVAR3": "VIIA3"
}

replacer = tickers_changes.get

tickers_skipped_updated = [replacer(n, n) for n in tickers_skipped]
```

# Updating tickers downloaded and tickers skipped list

```python
for ticker in tickers_skipped_updated:
  if ticker not in tickers_downloaded:
    print(f"Downloading {ticker}...")
    df = yf.download(str(ticker) + ".SA")[["Open", "Adj Close"]]
    if list(yf.shared._ERRORS.keys()):
      print(f"{ticker} not found, skipping...")
    else:
      df.to_csv(f"data/raw_data/{ticker}.csv")
      tickers_downloaded.append(ticker)
      downloaded += 1
  else:
    tickers_skipped_updated.remove(ticker)
```

```python
with open("data/tickers_skipped.pkl", "wb") as file:
    pickle.dump(tickers_skipped_updated, file)
```

# Importing data remaining from fundamentus (https://www.fundamentus.com.br/)

```python
def fund_data(ticker):
    url = f"https://fundamentus.com.br/amline/cot_hist.php?papel={ticker}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0",
        "Referer": "https://fundamentus.com.br/cotacoes.php?"
    }
    response = requests.request("GET", url, headers=headers)
    index = []
    adj_close = []
    for i in response.text.split("\n")[1:-2]:
        index.append(pd.to_datetime(i.split(",")[1][1:], unit="ms"))
        adj_close.append(float(i.split(",")[2][:-2]))
    pd.DataFrame(adj_close, index=index, columns=["Adj Close"]).to_csv(f"data/raw_data_fund/{ticker}.csv")

    return      
```

```python
for ticker in tickers_skipped:
  fund_data(ticker)
```
